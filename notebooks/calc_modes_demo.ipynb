# Calc Modes Demo

Demonstrates the completed `arb_calc` and `acb_calc` functions with the four modes:

- `point`
- `basic`
- `adaptive`
- `rigorous`


```python
import jax.numpy as jnp

from arbplusjax import acb_core, api, calc_wrappers, double_interval as di
```

```python
def iv(lo, hi):
    return di.interval(jnp.float64(lo), jnp.float64(hi))

def box(re_lo, re_hi, im_lo, im_hi):
    return acb_core.acb_box(iv(re_lo, re_hi), iv(im_lo, im_hi))

a = iv(-0.2, 0.1)
b = iv(0.7, 0.9)
z0 = box(-0.2, 0.1, -0.05, 0.05)
z1 = box(0.7, 0.9, 0.1, 0.2)
```

## Point Mode via API

```python
api.eval_point("arb_calc_partition", a, b, 6)
```

## Interval Modes via Generated Calc Wrappers

```python
for mode in ("basic", "adaptive", "rigorous"):
    out = calc_wrappers.arb_calc_partition_mode(a, b, 6, impl=mode, dps=50)
    print(mode, out.shape, out[:2])
```

```python
for mode in ("basic", "adaptive", "rigorous"):
    out = calc_wrappers.acb_calc_integrate_mode(z0, z1, impl=mode, dps=50)
    print(mode, out)
```

## New Calc APIs

```python
print("arb_calc_newton_step", api.eval_interval("arb_calc_newton_step", iv(0.2, 0.4), mode="basic", dps=50))
print("arb_calc_refine_root_newton", api.eval_interval("arb_calc_refine_root_newton", iv(0.2, 0.4), mode="basic", dps=50))
print("arb_calc_isolate_roots", api.eval_interval("arb_calc_isolate_roots", a, b, 6, mode="basic", dps=50).shape)
print("acb_calc_cauchy_bound", api.eval_interval("acb_calc_cauchy_bound", z0, z1, mode="basic", dps=50))
print("acb_calc_integrate_gl_auto_deg", api.eval_interval("acb_calc_integrate_gl_auto_deg", z0, z1, mode="basic", dps=50))
print("acb_calc_integrate_taylor", api.eval_interval("acb_calc_integrate_taylor", z0, z1, mode="basic", dps=50))
```
